### Imports

```python
import pandas as pd
from sqlalchemy import create_engine
import pyodbc
```

### Read csv

```python
df = pd.read_csv('sample_data/03_Library SystemCustomers.csv')
```

### Drop null values

```python
df = df.dropna(subset='Customer ID')
```

### Standardise column names

```python
df.columns = df.columns.str.lower().str.replace(' ','_')
```

### Cast Integers

```python
df = df.astype({'customer_id': int})
```

### Output to csv

```python
df.to_csv('output_data/customers_output.csv', index=False)
```

### Output to SQL Server

```python
engine = create_engine('mssql+pyodbc://@localhost/m5library?driver=ODBC+Driver+17+for+SQL+Server')

df.to_sql('customers_output', engine, if_exists='replace', index=False)

#trying to get sql server login to work...
```
