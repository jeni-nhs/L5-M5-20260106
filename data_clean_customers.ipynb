### Imports

```python
import pandas as pd
```

### Read csv

```python
df = pd.read_csv('sample_data/03_Library SystemCustomers.csv')
```

### Drop null values

```python
df = df.dropna(subset='Customer ID')
```

### Cast Integers

```python
df = df.astype({'Customer ID': int})
```

### Output to csv

```python
df.to_csv('customers_output.csv', index=False)
```
