```python
import pandas as pd
```

### Read the csv

```python
df = pd.read_csv('sample_data/03_Library Systembook.csv')
```

### Remove NaN

```python
df = df.dropna()
```

### Standardise column names

```python
df.columns = df.columns.str.lower().str.replace(' ','_')
```

### Cast as ints

```python
df = df.astype({'id': int, 'customer_id': int})
```

### Remove leading and trailing spaces

```python
df['books'] = df['books'].str.strip()
```

### Remove quotation marks

```python
df['book_checkout'] = df['book_checkout'].str.replace('"','',regex=False)
```

### Add actual numbers

```python
df['days_allowed_to_borrow'] = df['days_allowed_to_borrow'].str.replace('2 weeks','14',regex=False).astype(int)
```

### Sort out titles

```python
df['books'] = df['books'].str.title()

mask = df['books'].str.startswith('Lord Of The Rings', na=False)
df.loc[mask,'books'] = df.loc[mask,'books'].str.replace('Kind','King',regex=False)
```

### Check for invalid dates and remove

```python
for col in ['book_checkout', 'book_returned']:
    df[col] = pd.to_datetime(df[col], dayfirst=True, errors='coerce')
```

```python
from checkout_days import checkout_days

df['days_borrowed'] = df.apply(lambda row: checkout_days(row['book_checkout'],row['book_returned']), axis=1).astype('Int64')
```

### Output to csv

```python
df.to_csv('output_data/books_output.csv', index=False)
```
