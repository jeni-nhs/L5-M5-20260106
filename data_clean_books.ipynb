```python
import pandas as pd
```

### Read the csv

```python
df = pd.read_csv('sample_data/03_Library Systembook.csv')
```

### Remove NaN

```python
df = df.dropna()
```

### Cast as ints

```python
df = df.astype({'Id': int, 'Customer ID': int})
```

### Remove leading and trailing spaces

```python
df['Books'] = df['Books'].str.strip()
```

### Remove quotation marks

```python
df['Book checkout'] = df['Book checkout'].str.replace('"','',regex=False)
```

### Add actual numbers

```python
df['Days allowed to borrow'] = df['Days allowed to borrow'].str.replace('2 weeks','14',regex=False).astype(int)
```

### Output to csv

```python
df.to_csv('output_data/books_output.csv', index=False)
```
